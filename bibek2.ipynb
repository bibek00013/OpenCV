Real time face detection

```python
import numpy as np 
import matplotlib.pyplot as plt
import cv2
```

Accessing the webcam

```python
cap = cv2.VideoCapture(0)
```

```python
#loading haar_cascacde model

face_cascade = cv2.CascadeClassifier(
    "haarcascade_frontalface_default.xml"
)

if face_cascade.empty():
    print("Error in loading cascade")

while True:
    ret, frame = cap.read() #Reads one frame from the webcam.
    if not ret:
        break
    gray = cv2.cvtColor(frame , cv2.COLOR_BGR2GRAY)
    
    faces = face_cascade.detectMultiScale(
        gray, 
        scaleFactor= 1.1,
        minNeighbors=5,
        minSize=(30,30)
    )

    for (x,y,w,h) in faces:
        cv2.rectangle(
            frame,
            (x,y),
            (x+h, y+w),
             (0,255,0),
             2)
        
    cv2.imshow("Face detection ",frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
    #Waits 1 millisecond for a key press.
    #Returns an integer key code.
    #This value may contain extra bits depending on the OS

cap.release()
cv2.destroyAllWindows()
#Releases the camera.
#Closes all OpenCV windows.
```

0xFF
Keeps only the last 8 bits of the key code.

Removes unwanted higher-order bits.

Makes key detection consistent across platforms (Windows/Linux/Mac).
